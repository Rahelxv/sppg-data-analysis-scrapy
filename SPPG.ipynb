```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/sppg.csv')
df = df.dropna(how='all')
df = df.reset_index(drop=True)
df.tail()
```

# Jumlah provinsi yang punya sppg seluruh indonesia
semua provinsi punya

```python
df['Provinsi SPPG'] = df['Provinsi SPPG'].apply(
    lambda x: 'PAPUA' if "".join(str(x).split()) == 'PAPUA' else str(x).strip()
)
jumlah_provinsi_unique = df['Provinsi SPPG'].nunique()
list_provinsi =df['Provinsi SPPG'].unique()
print("Jumlah provinsi yang punya sppg: ", jumlah_provinsi_unique)
print("List provinsi yang punya sppg:")
for prov in list_provinsi:
  print(prov)
```

# Jumlah Kelurahan/Desa SPPG yang punya sppg

```python
# Menghitung jumlah desa unik yang punya SPPG
jumlah_desa_unique = df['Kelurahan/Desa SPPG'].nunique()

# Mengambil daftar nama DESA unik (bukan Kab./Kota)
list_desa_kelurahan = df['Kelurahan/Desa SPPG'].unique()

print(f"Jumlah desa/kelurahan yang punya SPPG: {jumlah_desa_unique}")
print("Daftar desa/kelurahan yang punya SPPG:")

for des in list_desa_kelurahan:
    print(f"- {des}")
```

# sppg jawa vs luar pulau jawa

around 56% penduduk indonesia di pulau jawa (https://databoks.katadata.co.id/demografi/statistik/4e9e52829d77f4b/lebih-dari-setengah-penduduk-indonesia-tinggal-di-jawa)

tapi 61% sppg di jawa sedangkan cuman 39% di luar jawa, menurut kepercayaan saya pembukaan sppg di luar jawa simply sulit karena aksesability.

```python
provinsi_jawa = [
    'BANTEN', 'JAWA BARAT', 'DKI JAKARTA', 'JAWA TENGAH',
    'DAERAH ISTIMEWA YOGYAKARTA', 'JAWA TIMUR'
]

df['Wilayah'] = df['Provinsi SPPG'].apply(
    lambda x: 'Pulau Jawa' if any(prov in str(x).upper() for prov in provinsi_jawa) else 'Luar Pulau Jawa'
)

perbandingan = df['Wilayah'].value_counts()
persentase = df['Wilayah'].value_counts(normalize=True) * 100

print("Jumlah Unit SPPG:")
print(perbandingan)
print("\nPersentase Distribusi:")
print(persentase)
```
